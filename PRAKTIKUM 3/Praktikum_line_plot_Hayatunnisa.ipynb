**Multiple Line Plot & Customisasi**

```python
import matplotlib.pyplot as plt

#data sample
months = ['Jan', 'Feb', 'Mar', 'Apr', 'Mai', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
product_A_sales = [10, 20, 15, 25, 30, 45, 40, 50, 60, 55, 65, 70]
product_B_sales = [5, 10, 8, 15, 18, 20, 22, 30, 25, 35, 40, 45]

#line plot
#kustomisasi line produk a
plt.plot(months, product_A_sales, label='Product A', linestyle="--", marker='o', color='green')
#kustomisasi line produk b
plt.plot(months, product_B_sales, label='Product B', linestyle="-", marker='x', color='blue')

plt.title('Penjualan produk per bulan')
plt.xlabel('Bulan')
plt.ylabel('Penjualan')
plt.legend()
plt.grid(True)
plt.show()
```

**Jenis Garis untuk Menunjukkan Tren**

```python
import matplotlib.pyplot as plt

#data sample
months = ['Jan', 'Feb', 'Mar', 'Apr', 'Mai', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
product_A_sales = [10, 20, 15, 25, 30, 45, 40, 50, 60, 55, 65, 70]
product_B_sales = [5, 10, 8, 15, 18, 20, 22, 30, 25, 35, 40, 45]
product_C_sales = [18, 2, 25, 28, 32, 38, 42, 48, 52, 56, 60, 64]
product_D_sales = [7, 9, 11, 13, 16, 18, 20, 23, 25, 28, 30, 33]


#PLOT PRODUK A
plt.plot(months, product_A_sales, label='Product A', linestyle="--", color='green')

#PLOT PRODUK B
plt.plot(months, product_B_sales, label='Product B', linestyle="-", color='orange')

#PLOT PRODUK C
plt.plot(months, product_C_sales, label='Product C', linestyle=":", color='red')

#PLOT PRODUK D
plt.plot(months, product_D_sales, label='Product D', linestyle="-.", color='blue')

plt.title('Penjualan produk per bulan')
plt.xlabel('Bulan')
plt.ylabel('Jumlah Penjualan')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

#buat data sample (copied from previous cells to make this cell self-contained)
months = ['Jan', 'Feb', 'Mar', 'Apr', 'mei', 'jun', 'jul', 'Aug', 'Sep', 'Okt', 'Nov', 'Des']
product_C_sales = [18,2,25,28,32,38,42,48,52,56,60, 64] # 11 elements
product_D_sales = [7,9,11, 13,16,18,20,23,25,28,30,33] # 12 elements

#subplot, 2 baris 1 kolom, panjang 10 inch, tinggi 8 inch
fig, axs = plt.subplots(2, 1, figsize=(10,8))

#plot pertama untuk produk C
axs[0].plot(months, product_C_sales, label='Product C', color='red', marker='d')
axs[0].set_title('Penjualan Produk C per Bulan')
axs[0].set_xlabel('Bulan')
axs[0].set_ylabel('Jumlah Penjualan')
axs[0].legend()
axs[0].grid(True)

#plot kedua untuk produk D
axs[1].plot(months, product_D_sales, label='Product D', color='blue', marker='s') # Corrected to plot product_D_sales
axs[1].set_title('Penjualan Produk D per Bulan')
axs[1].set_xlabel('Bulan')
axs[1].set_ylabel('Jumlah Penjualan')
axs[1].legend()
axs[1].grid(True)

plt.tight_layout()
plt.show()
```

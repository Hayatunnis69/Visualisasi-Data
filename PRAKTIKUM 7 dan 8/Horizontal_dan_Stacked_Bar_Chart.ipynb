# **Import Library**

```python
import matplotlib.pyplot as plt
import numpy as np
```

# **Dataset**

```python
#data penjualan

brands = ['Samsung', 'Iphone', 'Vivo', 'Oppo']
sales_2023 = [500, 700, 300, 400]
sales_2024 = [550, 750, 350, 450]
sales_2025 = [600, 800, 400, 500]
```

# **Basic Chart (Horizontal & Stacked Horizontal Bar Chart)**

```python
y_pos =np.arange(len(brands))

#canvas dan axis grafik
fig1, ax1, = plt.subplots()

#grafik batang horizontal
ax1.set_title('Horizontal Bar Chart')
ax1.set_xlabel('Jumlah Penjualan')
ax1.set_ylabel('Merk')
ax1.barh(brands, sales_2023)
plt.show()

#stacked horizontal bar chart
fig2, ax2, = plt.subplots() #canvas dan axis grafik

#grafik batang horizontal
ax2.set_title('Stacked Horizontal Bar Chart - 2023 & 2025')
ax2.set_xlabel('Jumlah Penjualan')
ax2.set_ylabel('Merk')

#bar pertama untuk 2023
ax2.barh(brands, sales_2023, label = '2023')

#bar kedua untuk 2024
ax2.barh(brands, sales_2024, left=sales_2023, label = '2024')

#legend
ax2.legend()
plt.show()
```

# **Kustomisasi Grafik**

```python
y_pos =np.arange(len(brands))
colors = ['Pink', 'skyblue', 'lightgreen', 'orange']

#custom horizontal bar chart
fig3, ax3, = plt.subplots()

ax3.barh(brands, sales_2023, color = colors, edgecolor = 'black', height=0.6)
ax3.set_title('Stacked Horizontal Bar Chart - 2023 & 2025')
ax3.set_xlabel('Jumlah Penjualan')
ax3.set_ylabel('Merk')
ax3.grid(axis='x', linestyle='--', alpha=0.7)

#menampilkan label nilai di bar chart
for i, v in enumerate(sales_2023):
  ax3.text(v+10, i, str(v), color='black', fontweight='bold')

plt.show()

#custom stacked horizonal bar chart (2023-2024)
fig4, ax4, = plt.subplots()

ax4.barh(brands, sales_2023, color = 'Maroon', edgecolor = 'black', label = '2023')
ax4.barh(brands, sales_2024, left=sales_2023, color = 'yellow', edgecolor = 'black', label = '2024')


ax4.set_title('Penjualan Smartphone Tahun 2023 & 2024 (Custom)')
ax4.set_xlabel('Jumlah Penjualan')
ax4.set_ylabel('Merk')
ax4.grid(axis='x', linestyle='--', alpha=0.5)

#hitung total penjualan dari 2 tahun untuk label angka
total_sales = np.array(sales_2023) + np.array(sales_2024)

#label nilai
for i, v in enumerate(total_sales):
  ax4.text(v+10, i, str(v), va = 'center')

ax4.legend()
plt.show()
```

# **Multiple Chart**

```python
y_pos =np.arange(len(brands))

fig5, ax5 = plt.subplots()

#bar pertama
ax5.barh(y_pos - 0.2, sales_2023, color = 'yellow', height=0.4, label = '2023')

#bar kedua
ax5.barh(y_pos + 0.2, sales_2024, color = 'lightblue', height=0.4, label = '2024')

ax5.set_yticks(y_pos, brands)
ax5.set_title('Penjualan Smartphone Tahun 2023 vs 2024 (Custom)')
ax5.set_xlabel('Jumlah Penjualan')
ax5.set_ylabel('Merk')
ax5.legend()
plt.show()

#multiple stacked horizontal bar chart
fig6, ax6 = plt.subplots()

#bar pertama
ax6.barh(y_pos, sales_2023, color = 'pink', label = '2023')

#bar kedua
ax6.barh(y_pos, sales_2024, left = sales_2023, color = 'lightgreen', label = '2024')

#bar ketiga
ax6.barh(y_pos, sales_2025, left=np.array(sales_2023) + np.array(sales_2024), color = 'lightblue', label = '2025')

ax6.set_yticks(y_pos, brands)
ax6.set_title('Penjualan Smartphone Tahun 2023 vs 2024 vs 2025 (Custom)')
ax6.set_xlabel('Jumlah Penjualan')
ax6.set_ylabel('Merk')
ax6.legend()
plt.show()
```

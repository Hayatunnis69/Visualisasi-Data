**Basic Stacked Vertical Bar Chart**

```python
import matplotlib.pyplot as plt
import numpy as np

# Dataset definitions (added for self-containment)
stores = ['Store A', 'Store B', 'Store C']
male = [150, 200, 180]
female = [120, 230, 170]

x = np.arange(len(stores))
plt.bar(x, male, label = 'Male', color = 'lightblue')
plt.bar(x, female, bottom=male, label = 'Female', color = 'Pink')

plt.xlabel('Stores')
plt.ylabel('Population')
plt.title('Population by Gender and Store')
plt.xticks(x, stores) # Added store names for x-axis labels
plt.legend()

plt.show()
```

**Kustomisasi Stacked Vertical Bar Chart**

```python
import numpy as np
import matplotlib.pyplot as plt

# Dataset definitions (added for self-containment)
stores = ['Store A', 'Store B', 'Store C']
male = [150, 200, 180]
female = [120, 230, 170]

x = np.arange(len(stores))
plt.bar(x, male, label = 'Male', color = 'lightblue')
plt.bar(x, female, bottom=male, label = 'Female', color = 'Pink')

plt.xlabel('Stores')
plt.ylabel('Population')
plt.title('Population by Gender and Store')
plt.xticks(x, stores) # Added store names for x-axis labels
plt.legend()

for i in range(len(x)):
    plt.text(x[i], male[i]/2, str(male[i]), ha = 'center', color = 'white')
    plt.text(x[i], male[i] + female[i]/2, str(female[i]), ha = 'center', color = 'black')

plt.show()
```

**Multiple Stacked Vertical Bar Chart**

```python
import matplotlib.pyplot as plt
import numpy as np

bar_width = 0.2
x = np.arange(len(stores))

#quarter 1
plt.bar(x - bar_width * 1.5, q1_male, label = 'Q1 Male', color = 'lightblue', width=bar_width)
plt.bar(x - bar_width * 1.5, q1_female, bottom=q1_male, label = 'Q1 Female', color = 'pink', width=bar_width)


#quarter 2
plt.bar(x - bar_width / 2, q2_male, label = 'Q2 Male', color = 'navy', width=bar_width)
plt.bar(x - bar_width / 2, q2_female, bottom=q2_male, label = 'Q2 Female', color = 'yellow', width=bar_width)


#quarter 3
plt.bar(x + bar_width / 2, q3_male, label = 'Q3 Male', color = 'chocolate', width=bar_width)
plt.bar(x + bar_width / 2, q3_female, bottom=q3_male, label = 'Q3 Female', color = 'green', width=bar_width)


#quarter4
plt.bar(x + bar_width * 1.5, q4_male, label = 'Q4 Male', color = 'red', width=bar_width)
plt.bar(x + bar_width * 1.5, q4_female, bottom=q4_male, label = 'Q4 Female', color = 'maroon', width=bar_width)


plt.title('Population by Gender and Store (Multiple Quarter)')
plt.xlabel('Stores')
plt.ylabel('Population')
plt.xticks(x, stores)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

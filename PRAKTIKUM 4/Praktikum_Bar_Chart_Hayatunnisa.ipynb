**Basic Bar Chart**

```python
import matplotlib.pyplot as plt

#data
jurusan = [
    'Ilmu Komputer',
    'Teknik Informatika',
    'Sistem Informasi',
    'Data Science'
]

jumlah_mahasiswa = [120, 150, 100, 80]


#membuat basic bar chart
plt.bar(jurusan, jumlah_mahasiswa, color=['maroon', 'blue', 'yellow', 'green'])
plt.xlabel('Jurusan')
plt.ylabel('Jumlah Mahasiswa')
plt.title('Jumlah Mahasiswa Per Jurusan')
plt.show()
```

**Basic Bar Chart dengan Matplotlib**

```python
import matplotlib.pyplot as plt

#data
jurusan = [
    'Ilmu Komputer',
    'Teknik Informatika',
    'Sistem Informasi',
    'Data Science'
]

jumlah_mahasiswa = [120, 150, 100, 80]


#membuat basic bar chart
plt.bar(jurusan, jumlah_mahasiswa, color=['pink', 'skyblue', 'yellow', 'chocolate'])
plt.xlabel('Jurusan')
plt.ylabel('Jumlah Mahasiswa')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Customisasi Basic Bar Chart

```python
import matplotlib.pyplot as plt

#data
jurusan = [
    'Ilmu Komputer',
    'Teknik Informatika',
    'Sistem Informasi',
    'Data Science'
]

jumlah_mahasiswa = [120, 150, 100, 80]


#membuat basic bar chart
plt.bar(jurusan, jumlah_mahasiswa, color=['orange', 'pink', 'yellow', 'skyblue'])
plt.xlabel('Jurusan')
plt.ylabel('Jumlah Mahasiswa')
plt.xticks(rotation=45)

#menampilkan label pada batang
for i, v in enumerate(jumlah_mahasiswa):
    plt.text(i, v, str(v), ha='center')

plt.show()
```

**Multiple Bar Chart**

```python
import matplotlib.pyplot as plt

#data
jurusan = [
    'Ilmu Komputer',
    'Teknik Informatika',
    'Sistem Informasi',
    'Data Science'
]

jumlah_mahasiswa = [120, 150, 100, 80]
jumlah_mahasiswa_2024 = [110, 140, 95, 85]


x = range(len(jurusan))
width = 0.4

plt.bar(x, jumlah_mahasiswa, width, label='2023', color='green')
plt.bar([p + width for p in x], jumlah_mahasiswa_2024, width=width,
        label='2024', color='Maroon')


plt.title('Jumlah Mahasiswa Per Jurusan (2023 VS 2024)')
plt.xlabel('Jurusan')
plt.ylabel('Jumlah Mahasiswa')
plt.xticks(rotation=45)
plt.xticks([p + width/2 for p in x], jurusan)
plt.legend()


plt.show()
```

**Basic Bar Chart untuk Melihat Pola dan Tren**

```python
import matplotlib.pyplot as plt

#data
jurusan = ['Ilmu Komputer', 'Teknik Informatika', 'Sistem Informasi', 'Data Science']
tahun = ['2019', '2020', '2021', '2022', '2023']
data = {
    'Ilmu Komputer': [100, 110, 120, 130, 140],
    'Teknik Informatika': [120, 120, 132, 145, 160],
    'Sistem Informasi': [90, 95, 100, 105, 110],
    'Data Science': [70, 75, 80, 85, 90]
}

width = 0.2
x = range(len(tahun))

#visualisasi masing-masing jurusan
plt.bar(x, data['Ilmu Komputer'], width=width, label='Ilmu Komputer', color='pink')
plt.bar([p + width for p in x],
        data['Sistem Informasi'], width=width, label='Sistem Informasi', color='orange')
plt.bar([p + 2 * width for p in x],
        data['Teknik Informatika'], width=width, label='Teknik Informatika', color='green')
plt.bar([p + 3 * width for p in x],
        data['Data Science'], width=width, label='Data Science', color='skyblue')

#detail grafik
plt.title('Jumlah Mahasiswa Per Jurusan (2019 vs 2023)')
plt.xlabel('Tahun')
plt.ylabel('Jumlah Mahasiswa')
plt.xticks([p + width for p in x], tahun)
plt.legend()

plt.show()
```
